The `mmr` column has holds ðŸ¤’ measles, mumps, rubella ðŸ¤’ vaccination rates for students in each respective school. **If the `mmr` value is not available for a school the `mmr` value is set to -1 in this data set.** 

The target `mmr` vaccination rate as [recommended by the CDC](https://www.cdc.gov/mmwr/preview/mmwrhtml/mm5632a3.htm) for schools is 95%. 

**Calculate the fraction of schools per state that have vaccination rates greater than the CDC 95%. Capture the output as a table called `df_vacc_rates`.**

You'll need to do use `filter`, `group_by`, and `summarize`. 

I.e.
1. Filter out schools that don't report `mmr` vaccination rate (keep schools where `mmr >= 0`).
1. Group the data by `state`.
1. Summarize fraction of vaccination rates above 95%.

ðŸ’¡Remember `n()` stands for the number or records in a group. Also, `sum(mmr >= 95, na.rm=TRUE)` will count the number of values above or equal to 95.ðŸ’¡

```r
# create the df_vacc_rates here

df_vacc_rates = measles |>
                filter(mmr >= 0) |>
                group_by(state) |>
                summarize(frac = sum(mmr>=95, na.rm = TRUE)/n())

df_vacc_rates |> head()
```

Which state (of those that report `mmr`) has the smallest fraction of schools above the CDC target vaccination rate of 95%?  

```r
df_vacc_rates |> 
    slice_min(frac)
```

**Make a ECDF plot of the school vaccination rates in North Carolina.** Use the `overall` column which reports the "overall" vaccination rate at each school.

â—ï¸Remember, you can calculate the `y` value for ECDF charts using `row_number`, `mutate`, and `arrange`.

```r
ecdf = measles |>
    filter(state == 'North Carolina') |>
    filter(overall >= 0) |>
    arrange(overall) |>
    group_by(state) |>
    mutate(rn = row_number(), fraction_overall = rn / max(rn))


ecdf|> head()
```

```r
vline = data.frame(
        vline95 = c('CDC recommended \n vaccination \n rate (95%)'),
        overall = (c(x = 95)), fraction_overall = (c(y=0.7)))

vline |> head()
```

```r
p = ggplot(ecdf, aes(x = overall, y = fraction_overall)) +
    geom_point(color = '#4B9CD3') + 
    geom_line(color = '#4B9CD3') +
    geom_vline(xintercept = 95, color = '#87A5C4', linetype = "dashed") +
    geom_text_repel(
        data = vline,
        aes(label = vline95, x = overall, y = fraction_overall),
        hjust = 0.5,
        nudge_x = 0.5,
        nudge_y = 0.05,
        box.padding = 1.55,
        segment.curvature = 0.5,
        segment.ncp = 1,
        segment.angle = 0.7,
        arrow = arrow(length = unit(0.015, "npc"))
    ) +
    labs(y = 'Fraction of Schools', x = 'Vaccination Rate(%)', color = '') + 
    theme_minimal() +
    theme(legend.position = 'top', legend.text = element_text(face = 'italic'))

p
```
