Use `geom_bar` to make a histogram of penguin counts by `year`.

```r
options(repr.plot.width=2, repr.plot.height=2)
p = ggplot(penguins, aes(x = year)) + 
    geom_bar(color = "#7BAFD4", fill = "#7BAFD4") + 
    scale_y_continuous(labels = scales::comma_format()) + 
    labs(x = "Year", y = "Count") + 
    ggtitle("Count of penguins by year")

p
```

Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.width=4, repr.plot.height=4)
p = ggplot(penguins, aes(x = body_mass_g)) + 
    geom_histogram(color = "#7BAFD4", fill = "#7BAFD4", alpha = 0.25, na.rm = TRUE) + 
    scale_x_continuous(labels = scales::comma_format()) + 
    labs(x = "Body Mass in grams", y = "number of penguins") + 
    facet_wrap(~year, ncol = 1, strip.position="right") +
    theme_minimal() +  
    ggtitle("Distribution of body mass by year")

p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.width=4, repr.plot.height=4)
p = ggplot(penguins, aes(x = body_mass_g, y = sex)) + 
    geom_point(na.rm = T, size = 1, alpha = 0.5, color = '#7BAFD4', position = position_jitter(h = 0.2, w = 0.3)) + 
    labs(x = "Body Mass in grams", y = "") + 
    facet_wrap(~species, scales = 'free_y', strip.position="top", ncol = 1) +
    scale_x_continuous(labels = scales::comma_format()) +
    theme_minimal() +  
    ggtitle("Body Mass Male vs Female by species")

p
```
