````r
gs4_deauth() # won't get prompted for a token
````

````r
salary = read_sheet('https://docs.google.com/spreadsheets/d/1G0FmJhkOME_sv66hWmhnZS5qR2KMTY7nzkxksv46bfk/edit#gid=203339396')
````

````r
salary |> head()
````

````r
options(repr.plot.height = 3)

p = ggplot(salary, aes(x = `Years of Experience`, y = `Compensation - Annual Base Salary/Pay`)) +
        geom_point(alpha = 0.5) +
        geom_smooth() +
    theme_minimal() +
    scale_y_continuous(labels = scales::dollar_format())
        
p
````
