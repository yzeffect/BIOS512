```r
library('tidyverse')
library('patchwork')
```

```r
data_url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-01-11/colony.csv'

df_bees = read_csv(data_url, col_type = cols()) |>
    mutate(
        season = case_when(
            months == 'October-December' ~ 'winter',
            months == 'January-March' ~ 'winter',
            TRUE ~ 'summer'
        ),
    ) |>
    group_by(year, state, season)

df_bees |> head()
```

```r
options(repr.plot.height = 4, repr.plot.width = 8)

p = ggplot(df_bees, aes(x = year, y = colony_lost_pct)) + geom_point()

p
```

```r
options(repr.plot.height = 4, repr.plot.width = 8)

p = ggplot(df_bees, aes(x = year, y = colony_lost_pct)) + geom_point() + facet_wrap(~season)

p
```

```r
options(repr.plot.height = 4, repr.plot.width = 8)

p = ggplot(df_bees, aes(x = year, y = colony_lost_pct)) + geom_point() + facet_wrap(~season, ncol=1)

p
```

```r
options(repr.plot.height = 4, repr.plot.width = 8)

p = ggplot(df_bees, aes(x = year, y = colony_lost_pct)) + 
    geom_point(position = position_jitter(height = 0, width = 0.3), alpha = 0.5) + 
    facet_wrap(~season, ncol=1, scales = 'free')

# This helps you see all the data not on top of each other
# Alpha: Transparency. 1=opaque

p
```
