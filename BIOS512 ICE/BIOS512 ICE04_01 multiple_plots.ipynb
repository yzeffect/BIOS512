```r
options(repr.plot.width=10, repr.plot.height=4.5, repr.plot.res = 300)

pi = ggplot(iris, aes(x = Sepal.Length, y = Sepal.Width, color = Species)) + 
     geom_point() + 
     labs(x = "Sepal Length", y = "Sepal Width")

pm = ggplot(mpg, aes(x = displ, y = hwy)) + 
     geom_point() +
     labs(x = "Displacement", y = "Highway MPG")


pm+pi
```
