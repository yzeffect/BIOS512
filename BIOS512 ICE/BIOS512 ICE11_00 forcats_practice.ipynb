Make a bar chart of the frequency of values for one "categorical" column in the data you chose.

```r
dog2 = dog |> 
    group_by(`Coat Length`) |>
    mutate(coat_freq = n()) |>
    ungroup() |>
    arrange(desc(coat_freq)) 

p = ggplot(dog2, aes(y = `Coat Length`)) + geom_bar()
p
```
